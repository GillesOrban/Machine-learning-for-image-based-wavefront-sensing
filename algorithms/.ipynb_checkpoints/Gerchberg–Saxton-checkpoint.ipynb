```python
from astropy.io import fits
import numpy as np
import numpy.fft as FFT
import aotools
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def meterToRadiant(array, wavelength):
    array_rad = (array / (wavelength*(10**9))) * (2*np.pi)
    return array_rad

def getPhase(z_coeffs, z_basis):
    phase = np.squeeze(np.sum(zernike_coeffs[:, None, None] * zernike_basis[1:, :, :], axis=0))
    return phase

def fft(array):
    fft_array = FFT.fftshift(FFT.fft2(FFT.ifftshift(array)))
    return fft_array

def ifft(array):
    fft_array = FFT.ifftshift(FFT.ifft2(FFT.fftshift(array)))
    return fft_array

def pad_with(vector, pad_width, iaxis, kwargs):
    pad_value = kwargs.get('padder', 10)
    vector[:pad_width[0]] = pad_value
    vector[-pad_width[1]:] = pad_value
    return vector

def rmse(a, b):
    n = len(a)
    rms_error = np.sqrt(((a - b)**2).mean())
    return rms_error
```

```python
# Files name
reference_values = 'psf_ref.fits'
PointSpreadFunction =  'psf_0.fits'
# Files content
rv_HDU = fits.open(reference_values)
psf_HDU = fits.open(PointSpreadFunction)

# Wavelength in m
wavelength = 2.2e-6
# Pupil Mask
mask = rv_HDU[0].data
# Reference PSF (ie. without aberration)
psf_reference = rv_HDU[1].data
# Norm 
norm = np.amax(psf_reference)
# Aberrated PSF
psf = psf_HDU[1].data
# Corresponding Zernike coefficients
zernike_coeffs = psf_HDU[0].data
# Zernike polynomials
zernike_basis = aotools.zernikeArray(21, 128, norm='rms')
# Phase associated to 'zernike_coeffs'
phase = getPhase(zernike_coeffs, zernike_basis)

# Show data
f, axarr = plt.subplots(1,4, figsize=(18, 10))
im = axarr[0].imshow(mask)
axarr[0].set_title("Mask")
axarr[1].imshow(np.sqrt(psf_reference))
axarr[1].set_title("Reference PSF")
axarr[2].imshow(np.sqrt(psf))
axarr[2].set_title("Observed PSF")
axarr[3].imshow(phase)
axarr[3].set_title("Corresponding phase")
plt.show()
```

```python
# Simulated PSF

Zs = aotools.zernikeArray(20 + 1, 128, norm='rms')
coeffs = (np.random.random(20)-0.5) * 0.4
coeffs[0:2]=0
phase_ = np.squeeze(np.sum(coeffs[ :, None, None] * Zs[1:, :,:], axis=0))

mask = np.pad(mask, 256, pad_with, padder=0)
phase_ = np.pad(phase_, 256, pad_with, padder=0)

H = mask * np.exp(1j * phase_)
h = fft(H)
psf_test = np.abs(h)**2

# Show data
f, axarr = plt.subplots(1,2, figsize=(10, 10))
im = axarr[0].imshow(phase_[256:384,256:384])
axarr[0].set_title("Phase")
axarr[1].imshow(np.sqrt(psf_test[256:384,256:384]))
axarr[1].set_title("PSF")
plt.show()
```

```python
# Gerchberg-Saxton algorithm
# see pseudo code: wikipedia.org/wiki/Gerchberg-Saxton_algorithm
# -> Retrieve phase from psf (=target) and source (=input)

# Module
target = np.sqrt(np.abs(psf_test))
source = mask

# Iteration number
n_iteration = 100
errors = []

A  = ifft(target)
for i in range(n_iteration):
    B = np.absolute(source) * np.exp(1j * np.angle(A))
    C = fft(B)
    D = np.absolute(target) * np.exp(1j * np.angle(C))
    A = ifft(D)
    errors.append(rmse((mask*np.angle(A))[256:384,256:384], 
                        (mask*phase_)[256:384,256:384]))
    
    
retrieved_phase = mask * np.angle(A)

# Show phase
f, axarr = plt.subplots(1,2, figsize=(10, 10))
axarr[0].imshow(retrieved_phase[256:384,256:384])
axarr[0].set_title("GS phase")
axarr[1].imshow(phase_[256:384,256:384])
axarr[1].set_title("Real phase")
plt.show()
# Error
plt.plot(errors,'o-')
plt.title('RMSE')
plt.xlabel('iteration')
plt.ylabel('rmse')
plt.show()
```
