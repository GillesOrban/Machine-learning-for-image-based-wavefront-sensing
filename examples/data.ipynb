```python
# Project phase on Zernike Basis

# 1) Define the basis
z_basis = aotools.zernikeArray(101, 128, norm='rms')  # [rad]

# 2) Define Standard Dot product
def frobeniusInnerProduct(A, B):
    fIProduct = 0
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            fIProduct += A[i, j]*B[i, j]
    return fIProduct;

# 3) Project
id = 3
sample = dataset[id]
phase = sample['phase'].numpy()
image_in = sample['image'][0].numpy()
image_out = sample['image'][1].numpy()

z_norm = np.empty([100])
z_coeffs = np.empty([100])
for i in range(100):
    z_norm[i] = frobeniusInnerProduct(z_basis[i+1,:,:], z_basis[i+1,:,:])
    z_coeffs[i] = frobeniusInnerProduct(phase, z_basis[i+1,:,:]) / z_norm[i]
        
        
# 4) Plot first 60 coefficients
colors = ['#669f61','#ff884d','#2c3e50','#81b9c3','#bdc3c7']
nbr = 30;
i_zernike = np.arange(2, nbr+2)
f, ax = plt.subplots(2, 1, figsize=(15, 8))
w = 0.6
ax[0].plot(i_zernike-w/2, np.abs(z_coeffs[0:nbr]), linewidth=4*w, color=colors[2], zorder=3)
ax[0].bar(i_zernike-w/2, np.abs(z_coeffs[0:nbr]), width=w, label='Original', color=colors[4], zorder=3)
ax[0].set_ylabel('Absolute magnitude [Rad]')
ax[0].legend(fancybox=False, frameon=False)
ax[0].set_xticks(np.arange(2, nbr+2)[0::4])
ax[0].minorticks_on()
ax[1].plot(i_zernike-w/2, np.abs(z_coeffs[nbr:2*nbr]), linewidth=4*w, color=colors[2], zorder=3)
ax[1].bar(i_zernike-w/2, np.abs(z_coeffs[nbr:2*nbr]), width=w, label='Original', color=colors[4], zorder=3)
ax[1].set_ylabel('Absolute magnitude [Rad]')
ax[1].set_xlabel("Zernike indices")
ax[1].set_xticks(np.arange(2, nbr+2)[0::4])
ax[1].set_xticklabels(np.arange(nbr+2, nbr+nbr+2)[0::4])
ax[1].minorticks_on()
ax[0].set_ylim(0,0.45)
ax[1].set_ylim(0,0.45)
ax[0].minorticks_on()
ax[1].minorticks_on()
ax[0].grid(zorder=0, color='lightgray', linestyle='--', which = 'minor')
ax[1].grid(zorder=0, color='lightgray', linestyle='--', which = 'minor')
plt.show() 
```
